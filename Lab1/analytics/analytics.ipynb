```python
# Initialize a figure for the subplots

plt.figure(figsize=(18, 18))



# Colors for the lines

colors = ['blue', 'orange', 'green']



# Plot Execution Time vs. Number of Threads

plt.subplot(3, 1, 1)

for i, (name, dataset) in enumerate(datasets.items()):

    avg_execution_time = np.mean(dataset["Execution Time"], axis=0)

    plt.plot(dataset["Threads"], avg_execution_time, label=name, marker='o', color=colors[i])

plt.title("Execution Time vs. Number of Threads")

plt.xlabel("Number of Threads")

plt.ylabel("Execution Time (s)")

plt.grid()

plt.legend()



# Plot Speed-up vs. Number of Threads

plt.subplot(3, 1, 2)

for i, (name, dataset) in enumerate(datasets.items()):

    avg_speedup = np.mean(dataset["Speed-up"], axis=0)

    plt.plot(dataset["Threads"], avg_speedup, label=name, marker='o', color=colors[i])

plt.title("Speed-up vs. Number of Threads")

plt.xlabel("Number of Threads")

plt.ylabel("Speed-up")

plt.grid()

plt.legend()



# Plot Efficiency vs. Number of Threads

plt.subplot(3, 1, 3)

for i, (name, dataset) in enumerate(datasets.items()):

    avg_efficiency = np.mean(dataset["Efficiency (%)"], axis=0)

    plt.plot(dataset["Threads"], avg_efficiency, label=name, marker='o', color=colors[i])

plt.title("Efficiency vs. Number of Threads")

plt.xlabel("Number of Threads")

plt.ylabel("Efficiency (%)")

plt.grid()

plt.legend()



# Adjust layout and show the plots

plt.tight_layout()

plt.show()
```
